### Algoritmo Minimax

Este algoritmos utiliza una busqueda en profundidad limitada. Este algoritmo emplea una funcion estatica la cual devuelve valores en funcion del jugador maximo, este algortmo trata de maximizar el valor de los jugadores mientras que un oponente trata de minimizarlo. 


Dentro de la teoria de juegos este algoritmo sirve para minimizar la perdida que se puede esperar en un juego contra un adversario donde se tiene informacion necesaria para determinar esto. Este algoritmo permite determinar o elegir un movimiento en un escenario donde mi contrincate escoge lo peor. 

Pasos del algoritmo Minimax: 

1. Se Genera un arbol de juego, se genera los nodos hasta un estado final.
2. Calcula una funcion de utilidad para cada nodo fina.
3. Calcula el valor de los nodos superiores en base a los nodos inferiores. Para elegir los movimiento del jugador y oponente se eligen valores maximos y minimos representanto movimientos.
4. Realiza un movimiento en base al valor superio que tiene el nodo en el arbol de juego.
 
Para la optimizacion de la busqueda en el arborl Claude propuso que se utilizaran otras funciones heuristicas como la busqueda por profundidad o por tiempo de ejecucion. Este algoritmo garantiza encontrar un moviento asuminedo que el jugador siempre realiza los movimientos de manera optima, este algortmo dependera de su optimo desarrollo en base a la profundidad del arbol de juego generado. 

Una de las principales caracteristicas es de crear escenarios complicados empleando reglas sencillas. Los juegos con estos algoritmos pueden ser testeados con jugadores contra maquinas. Este algoritmo al ser recursivo su terminacion o corte esta condicionada, estas condiciones pueden ser:

    -Gana un jugador
    -Se ha explorado el arbol y no hay mas capas
    -Se ha exedido el tiempo
    -Se ha llegado a un nivel sin cambios.


### Alpha-Beta prunning

Alpha-Bta prunning es denominado una tecnica que sirve para reducir el numero de nodos que se tiene que evaluar por el algoritmo minimax. Este tipo de tecnica reliza un tipo de corte o poda del arbol de busqueda y los nodos que representan los movinenotos tales que permiten que al jugador el controncante no le supere.

Esta tecnica fue analizada por Donald Knuth y Ronald Moore en 1975 en donde prponian la eficiencia de esta tecnica que decia que es posible encontrar una decision o nodo correcto sin recorrer todo el arbol de buqueda. Esta tecnica produce un movimiento, o jugada similar al algoritmo minimax, sin embargo esta tecnica reduce el recorrido del arbol de busqueda y su eficiencia deoendera de como se vaya explorando el arbol y de sus nodos sucesores. 

Este metodo realiza un mejor rendimiento en cuanto al algortimo minimax en cuanto al tiempo empleodo para el calculo de un movimiento donde se explora las ramas mas favorables y se quita o no explora aquellas que tienen probabilidad de realizar una jugada errada.

La tecnica de ALpha-Beta prunning es muy aplicable en espacios grandes donde se requiere un rendiemiento y eficiencia, para ello se tiene que explorar aquelllas partes de arbol con mayor porbabilidad de exito. Dentro de esto tenemos 2 caracteristicas que son:

    -ALfa: Representa el valor maximo que se tiene en periodo o punto de eleccion a traves de la ruta de busqueda.
    -Beta: Esto representa el valor minimo o mas bajo encontrada en ese momento o punto a lo largo de la ruta de busqueda.

Estos valores de alfa y beta se van actualizando conforme se va avanzado la busqueda y se va realizando la poda de aquellas ramas del arbol que no representan un exito. Esto se realiza con un arbol y una busqueda heuristica con profundidad limitada para poder podar o itnerrunpir las busquedas en nodos no favorables. Este corte o poda no afecta al resultado final ya que se consigue el mismo resultado sin aplicar esta tecnica al algoritmo minimax.

def minimax(state, depth, player):

    if player == COMP:
        best = [-1, -1, -infinity]
    else:
        best = [-1, -1, +infinity]

    if depth == 0 or game_over(state):
        score = evaluate(state)
        return [-1, -1, score]

    for cell in empty_cells(state):
        x, y = cell[0], cell[1]
        state[x][y] = player
        score = minimax(state, depth - 1, -player)
        state[x][y] = 0
        score[0], score[1] = x, y

        if player == COMP:
            if score[2] > best[2]:
                best = score  # max value
        else:
            if score[2] < best[2]:
                best = score  # min value

    return best
